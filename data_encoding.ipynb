```python
import pandas as pd
```

```python
# Read with UTF-8 and write with CP949
with open("/content/CAvideos.csv", "r", encoding="utf-8", errors="replace") as f_in:
    content = f_in.read()

with open("/content/CAvideos_encoding.csv", "w", encoding="cp949", errors="replace") as f_out:
    f_out.write(content)
```

```python
df = pd.read_csv("/content/CAvideos_encoding.csv", encoding='cp949')
df.head()
```
